```python
import xgboost as xgb
from xgboost import plot_tree
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split

# -----------------------------------------------------
# 1. Membuat dataset sintetis untuk ilustrasi XGBoost
# -----------------------------------------------------
X, y = make_classification(
    n_samples=200,
    n_features=2,            # hanya 2 fitur agar pohon mudah dipahami
    n_informative=2,
    n_redundant=0,
    n_clusters_per_class=1,
    random_state=42
)

# Split data
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

# -----------------------------------------------------
# 2. Melatih Model XGBoost Classifier
# -----------------------------------------------------
model = xgb.XGBClassifier(
    n_estimators=1,        # hanya 1 tree agar visualisasi jelas
    max_depth=3,
    learning_rate=0.3,
    objective='binary:logistic',
    random_state=42
)

model.fit(X_train, y_train)


```

```python
# -----------------------------------------------------
# 3. Visualisasi Struktur Pohon XGBoost
# -----------------------------------------------------
plt.figure(figsize=(16, 12))
plot_tree(model, num_trees=0, rankdir='LR')  # LR = left to right
plt.title("Visualisasi Pohon Keputusan Model XGBoost", fontsize=16)
plt.show()
```
